```python
def print_okved_data(CODE_TO_FIND, okved=okved):
    '''
    На вход подается числовой код из трех колонок данных.
    Перебором проверяется, встречается ли такой код в ОКВЭД базе.
    
    Печатает раздел, подраздел и название, если код встречается и 
    возвращает пустое значение.
    Если код не находится, то возвращает код и строку "Не найдено"
    '''
    print('Искомый код:', '.'.join(map(str, CODE_TO_FIND)))
    for sector in okved.keys(): 
        for subsector in okved[sector].keys(): 
            for ok1 in okved[sector][subsector].keys():
                if format(CODE_TO_FIND[0], "02") == ok1:
                    for ok2 in okved[sector][subsector][ok1].keys():
                        if format(CODE_TO_FIND[1], "02") == ok2:
                            for ok3 in okved[sector][subsector][ok1][ok2].keys():
                                if format(CODE_TO_FIND[2], "02") == ok3:
                                    print(sector, 
                                          subsector, 
                                          okved[sector][subsector][ok1][ok2][ok3],
                                          sep='\n')
                                    return None
    return '.'.join(map(str, CODE_TO_FIND)), 'Не найдено'
```

```python
%%time
rand_row = random.randint(0, data.shape[0]-1)
rand_code = data.loc[rand_row, ['ok1', 'ok2', 'ok3']].head().values

print(f'row #{rand_row}')
print_okved_data(rand_code)
```
